Here's a quick example for data plotting

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
ax = sns.scatterplot(
    data = iris,
    x = 'sepal_length',
    y = 'sepal_width',
    hue = 'species',
    size = 'petal_length'
)

ax.legend(bbox_to_anchor=(1., 1.))

plt.show()
```
